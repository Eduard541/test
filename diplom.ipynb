```python
# создадим функцию  по построению графика точечной диаграммы техпараметров
def scatter_graph(col1, col2):
    # определим выбросы
    q75, q25 = np.percentile(df_big[col1].loc[0:],[75,25]) # величина первого и третьего квартилей
    intr_qr = q75 - q25  # межквартильный диапазон
    max_x = q75 + (1.5*intr_qr) # определение границы для выбросов
    
    q75, q25 = np.percentile(df_big[col2].loc[0:],[75,25]) # величина первого и третьего квартилей
    intr_qr = q75 - q25  # межквартильный диапазон
    max_y = q75 + (1.5*intr_qr) # определение границы для выбросов
    
    data  = df_big[(df_big[col1] < max_x) & (df_big[col2] < max_y)]
    
    # постронение графика 
    fig, ax = plt.subplots(figsize=(10,10))
    sns.scatterplot(x=data[col1], 
            y=data[col2],  hue = data['general_assessment'], ax=ax)
    ax.set_title(f'Точечая диаграмма распределения оценок параметров {col1} и {col2}', fontsize=12)
    ax.set_xlabel(f'{col1}', fontsize=12)
    ax.set_ylabel(f' {col2}', fontsize=12)
    plt.show()    
# Применим функцию и построим график
scatter_graph('Downlink Throughput(Kbps)','Video Streaming Download Throughput(Kbps)')
```

создадим функцию по переводу имени датафрейма в строковый формат(например df_good --> 'df_good') в качестве аргумента используем имя датафрейма (например df_good). Ниже менять будем только аргументы для этой функции.

Суть создания функции бутстреп в качестве аргумента - датафрейм и колонка с техпараметром. Создаем пустой список, куда будем записывать среднюю случайно сгенерированного массива из датафрейма и колонки с данными техпараметрами, генерируем массив размера, который соответствует датафрейму и колонка с техпараметром в качестве аргумента и так 3000 раз. В итоге получаем массив размером 3000 данных, состоящих из средних. Ниже менять будем только аргументы для этой функции.

```python
# создадим функцию бутстреп
def func(df, column):   # аргументы(имя DF, имя колонки техпараметра)
    m = []          # пустой список, куда будем записывать среднее значение сгенерированного случ.образом массива техпарметра
    for i in range(3000):  # число случайных генераций (3000 раз) техпараметра
        ram = random.choices(df[column].values, k = len(df[column])) # массив техпар-в, сгенр случ.образом, кол-во эл массива
        m1 = mean(ram) # средняя выборки техпараметра полученной случайным образом
        m.append(m1)  # добавление в список средних
    return(m) # возвращаем массив полученных средних
```

```python
# создадим функцию  по построению сводной таблицы
def func_table (data, data1, data2):
    '''
    функция (func_table) создает таблицу с левой и правой границей доверительного интервала средних сгенерированных 
    случайным образом средних значений техпараметра с использованием бутстрепа из одной выборки, тоже самое только из 
    второй выборки, доверительного интервала разниц этих средних, наличие (отсутствие) пересечения доверительных интервалов и 
    вхождение (не вхождение) нуля в доверительный интервал разниц средних
    - аргументы функции:
       - массив средних сгенерированных случайным образом средних значений техпараметра одной выборки
       - массив средних сгенерированных случайным образом средних значений техпараметра второй выборки
       - массив из разниц средних этих двух массивов
    - параметр один и тот же в обоих выборках
    - df_name_1 - имя одного датафрейма, df_name_2 - имя другого датафрейма, Параметр - сравниваемый параметр
    - уровень значимости 0.05
                
    '''    
    df3 = pd.DataFrame({'df_name_1': [], 'df_name_2': [], 'Параметр': [], 
                        'percentile_2.5 df_name_1 ': [], 'percentile_97.5 df_name_1': [], 
                        'percentile 2.5 df_name_2': [], 'percentile_97.5 df_name_2': [],
                        'пересечение интервалов': [],
                       'diff_mean_per_2.5': [], 'diff_mean_per_97.5': [], 
                        'вхождение 0 в довинтервал разницы средних': []})

    # перцентили (2.5 и 97.5)
    left_1 = np.percentile(array_of_good_means, 2.5)
    right_1 = np.percentile(array_of_good_means, 97.5)            
    left_2 = np.percentile(array_of_bad_means, 2.5)
    right_2 = np.percentile(array_of_bad_means, 97.5)
    
    # список из перцентилей (2.5 и 97.5) нужен для определения пересечения интервалов
    sp_left = [left_1, right_1]  
    sp_right = [left_2, right_2]
    
    # разница средних 
    diff_mean_left = np.percentile(mean_difference, 2.5)
    diff_mean_right = np.percentile(mean_difference, 97.5)
        
    # создание новой строки с данными
    new_row = {'df_name_1': g, 'df_name_2': b, 'Параметр': col, 
                        'percentile_2.5 df_name_1 ': left_1, 'percentile_97.5 df_name_1': right_1, 
                        'percentile 2.5 df_name_2': left_2, 'percentile_97.5 df_name_2': right_2,
                        'пересечение интервалов': min(sp_left[1], sp_right[1]) > max(sp_left[0], sp_right[0]),
                       'diff_mean_per_2.5': diff_mean_left, 'diff_mean_per_97.5': diff_mean_right, 
                        'вхождение 0 в довинтервал разницы средних': diff_mean_left < 0 < diff_mean_right}
        
    df3 = df3.append(new_row, ignore_index=True)    
    df3.style.highlight_min(subset = 'вхождение 0 в довинтервал разницы средних')
    df3 = df3.replace({'пересечение интервалов':{1:'Yes', 0:'No'}})
    df3 = df3.replace({'вхождение 0 в довинтервал разницы средних':{1:'Yes', 0:'No'}})
        
    return df3

# Применим функцию:
df_united = func_table(array_of_good_means, array_of_bad_means, mean_difference) # передадим аргументы в функцию
```

```python
# создадим функцию по построению графиков
def mean_graph(data, data1, data2):
    '''
    функция (mean_graph) предназначена для построения двух графиков:
    - гистограммы средних сгенерированных случайным образом средних значений техпараметра с использованием бутстрепа из 
      двух выборок с линиями (границы) доверительных интервалов
    - гистограмма разниц средних с линиями (границы) доверительного интервала
    
    - аргументы функции:
       - массив средних сгенерированных случайным образом средних значений техпараметра одной выборки
       - массив средних сгенерированных случайным образом средних значений техпараметра второй выборки
       - массив из разниц средних этих двух массивов
    параметр один и тот же в обоих выборках
    уровень значимости 0.05
            
    '''        
    fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(16, 10))

    # первый график: гистограмма средних значений
    axs[0].hist(x = array_of_good_means, bins = 'fd', color='r', alpha = 0.5, label = g)
    axs[0].hist(x = array_of_bad_means, bins = 'fd', color='g', alpha = 0.5, label = b)
    # доверительный интерал для good
    axs[0].axvline(np.percentile(array_of_good_means, 2.5), linestyle='--', color='b', 
                   label = (f'Доверительный интервал {g}'))
    axs[0].axvline(np.percentile(array_of_good_means, 97.5), linestyle='--', color='b')
    # доверительный интерал для bad
    axs[0].axvline(np.percentile(array_of_bad_means, 2.5), linestyle='--', color='m', 
                   label = (f'Доверительный интервал {b}'))
    axs[0].axvline(np.percentile(array_of_bad_means, 97.5), linestyle='--', color='m')
    axs[0].legend()    
    axs[0].grid()
    # заголовок:
    title = (f'Доверительные интервалы для средних значений {g} и {b} техпараметра {col}' )
    axs[0].set_title(title)
    # подписи по осям:
    axs[0].set_ylabel('Количество данных')
    axs[0].set_xlabel(f'Средние значение {col}')
    
    print(f'{g} 2.5% = {np.percentile(array_of_good_means, 2.5):.2f} 97.5% = {np.percentile(array_of_good_means, 97.5):.2f}')
    print(f'{b} 2.5% = {np.percentile(array_of_bad_means, 2.5):.2f} 97.5% = {np.percentile(array_of_bad_means, 97.5):.2f}')
    
    # второй график: гистограмма разница средних значений
    axs[1].hist(x = mean_difference, bins = 'fd')
    axs[1].axvline(np.percentile(mean_difference, 2.5), linestyle='-', color='b')
    axs[1].axvline(np.percentile(mean_difference, 97.5), linestyle='-', color='b')
    axs[1].grid()
    # заголовок:
    title = (f'Доверительные интервалы для разницы средних значений {g} и {b} техпараметра {col} ')
    axs[1].set_title(title)
    # подписи по осям:
    axs[1].set_ylabel('Количество данных')
    axs[1].set_xlabel(f'Разница средних значений {col}')
    
    plt.show()
    
# Применим функцию:
mean_graph(array_of_good_means, array_of_bad_means, mean_difference)
```

Подготовим данные в выборке df_good и df_bad_inter в параметрах "Downlink Throughput(Kbps)" и "Uplink Throughput(Kbps)" поскольку наличие выбросов может оказывать влияние на получение результата, то отфильтруем от выбросов параметры в выборках. Выбросы это то что лежит за границей усов (в боксплоте).

**Подготовим данные в выборке df_good в параметрах "Downlink Throughput(Kbps)" и "Uplink Throughput(Kbps)"**

```python
# определим границы начала выбросов в параметре "Downlink Throughput(Kbps)" в выборке df_good
q75, q25 = np.percentile(df_good['Downlink Throughput(Kbps)'].loc[0:],[75,25]) # величина первого и третьего квартилей
intr_qr = q75 - q25  # межквартильный диапазон
max_x = q75 + (1.5*intr_qr) # определение границы для выбросов
    
# определим границы начала выбросов в параметре "Uplink Throughput(Kbps)" от выбросов в выборке df_good    
q75, q25 = np.percentile(df_good['Uplink Throughput(Kbps)'].loc[0:],[75,25]) # величина первого и третьего квартилей
intr_qr = q75 - q25  # межквартильный диапазон
max_y = q75 + (1.5*intr_qr) # определение границы для выбросов
    
# отфильруем от выбросов параметры    
data_good  = df_good[(df_good['Downlink Throughput(Kbps)'] < max_x) & (df_good['Uplink Throughput(Kbps)'] < max_y)]
```

**Эмпирическое корреляционное отношение**

Определим тесноту связи одного  и того же параметра в разных выборках с разными оценками. В качестве выборки определим выборки, где есть разница в исследуемом параметре.

```python
sp_data1 = [data_good_9, data_bad_inter_1, data_bad_inter_2, data_bad_inter_3, data_bad_inter_5]
```

```python
# создадим функцию  по построению сводной таблицы
def func_emp_corel_relation (spisok_data, col):
    '''
    функция (func_emp_corel_relation) предназначена для расчета эмпирического корреляционного отношения для определения
    степени зависимости исследуемого параметра в подготовленных для этого выборках 
    - аргументы функции:
       - список выборок (датафремы)
       - исследуемый параметр в этих выборках
            
    '''    
    d1 = 0 # начальная стартовая сумма
    d2 = 0 # начальная стартовая сумма
    d3 = 0 # начальная стартовая сумма

    # блок для расчета 
    for data in sp_data1:
        data_array = data[col].values # определим массив
        var = np.var(data_array) # дисперсия массива
    
    # накопление суммы (общее количество данных во всех выборках)
        sum_len = len(data_array) + d1 # счетчик количества данных в массивах
        d1 = sum_len # 
    
    # накопление данных (сумма произведений дисперсий на количество данных в этом массиве)
        tt = var* len(data) # 
        sum_tt = tt + d2 # счетчик сумм произведений дисперсий на количество данных в этом массиве
        d2 = sum_tt
 
    # накопление суммы 
        sum_mean = mean(data_array)*len(data) + d3
        d3 = sum_mean

    mean_vn = sum_tt/sum_len # Определим среднюю из внутригрупповых дисперсий
    mean_s = sum_mean/sum_len # Определим общую среднюю величину для расчета междугрупповой дисперсии:
    
    # блок для расчета междугрупповой дисперсии
    d4 = 0 # начальная стартовая сумма
    for data in sp_data1:
        data_array = data[col].values # определим массив
    
    # накопление суммы
        sum_mean_t = (mean(data_array) - mean_s)**2 * len(data) + d4 # счетчик сумм
        d4 = sum_mean_t
    
    meg_mean = sum_mean_t/sum_len # Определим междугрупповую дисперсию

    ddd = sqrt(meg_mean/(mean_vn + meg_mean)) # величина эмпирического корреляционного отношения    
    return ddd

# Применим функцию:
func_emp_corel_relation(sp_data1, 'Downlink Throughput(Kbps)') # передадим аргументы в функцию
```
